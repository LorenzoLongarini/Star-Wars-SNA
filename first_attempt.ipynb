```python
import json

path = 'db/starwars-full-interactions-allCharacters.json'
path1 = 'db/starwars-episode-1-interactions-allCharacters.json'
# Carica i dati JSON
with open(path1, 'r') as f:
    data = json.load(f)
```

```python
import networkx as nx

# Creazione di un grafo non diretto
G = nx.Graph()

# Aggiungi nodi
for node in data['nodes']:
    G.add_node(node['name'], value=node['value'], colour=node['colour'])
   

# Aggiungi archi
for link in data['links']:
    source = data['nodes'][link['source']]['name']
    target = data['nodes'][link['target']]['name']
    G.add_edge(source, target, value=link['value'])
```

```python
list(G.nodes)
```

```python
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)
plt.figure(figsize=(20,10))
nx.draw(G, pos, with_labels=True)
plt.show()
```

```python
# Calcolo della centralit dei nodi
centrality = nx.degree_centrality(G)

# Calcolo della clustering coefficient
clustering_coefficient = nx.clustering(G)

# Trova i nodi pi importanti
top_nodes = sorted(centrality, key=centrality.get, reverse=True)[:10]

# Stampa i nodi pi importanti
print("Top 10 nodes by centrality:")
for node in top_nodes:
    print(f"{node}: {centrality[node]}")
```
